```python
import pandas as pd
import numpy as np
import psycopg2
```

```python
# 2.1. Вызывать запрос из SQL базы demo_medium 
# Берём запрос из промежуточной аттестации, вьюшка в базе уже создана и сохранена 

query = '''
    SELECT 
        ROW_NUMBER() OVER() AS "№", 
        dep_airport_name AS "Аэропорт вылета", 
        arr_airport_name AS "Аэропорт прилёта", 
        fly_month AS "Месяц, год", 
        count_tickets AS "Количество билетов" 
    FROM view_flights 
    WHERE dep_airport_code = 'DME' 
    '''
engine = psycopg2.connect("dbname='demo' user='postgres' host='localhost' port='5432' password='123'") 

df = pd.read_sql(query, con=engine) 
```

```python
# 2.5. Считать частоту встречающихся значений по строковой переменной 

df['Аэропорт прилёта'].value_counts().sort_index() 

# Ещё пришлось отсортировать для красоты 
# Каждый аэропорт прилёта повторяется три раза, т.к. выборка у нас за три летних месяца 
```

```python
# 2.6. Считать минимум, среднюю, медиану, максимум по числовой переменной 
# Чтобы не изобретать велосипед, возьмём готовую функцию из консультации 09.09.2024 

def get_numeric_stats(df, columns_to_exclude = []):
    data_types = df.dtypes
    numeric_columns = data_types.loc[(data_types == int)|(data_types == float)].index.to_list()
    for col in columns_to_exclude:
        numeric_columns.remove(col)
    num_stats = df[numeric_columns].describe().loc[['min', 'mean', 'max']]
    return num_stats

print(get_numeric_stats(df, ['№'])) # Исключаем столбец '№' 
```
